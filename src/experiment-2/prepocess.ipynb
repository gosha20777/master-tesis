```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
def plt_display(image, title):
    fig = plt.figure()
    a = fig.add_subplot(1, 1, 1)
    imgplot = plt.imshow(image)
    a.set_title(title)
```
